*Star Pattern printing cross odd length numbers*

```python
nums = "1234567"

for i in range(len(nums)):
    for j in range(len(nums)):
        if i == j or i + j == len(nums)-1:
            print(nums[i], end = " ")
        else:
            print(" ", end = " ")
    print()
```

*Pattern printing numbers*

```python
nums = 5

for i in range(nums):
    k = 111
    for j in range(nums-i-1):
        print("   ", end = " ")
    for j in range(i+1):
        print(k, end =" ")
        k += 111
    print()
        
```

*Pattern printing numbers*

```python
nums = 5

for i in range(nums):
    k = 111
    for j in range(nums-i-1):
        print("   ", end = " ")
    k *= (i+1)
    for j in range(i+1):
        print(k, end =" ")
        k -= 111
    print()
```

LCM

```python
class Solution:
    # @staticmethod
    def Compute_LCM(self,n1, n2):
        if n1 > n2:
            higher = n1
        else:
            higher = n2

        value = higher

        while True:
            if higher % n1 == 0 and higher % n2 == 0:
                return(f"LCM of {n1} and {n2} is {higher}")
                break
            else:
                higher += value
sol = Solution()
print(sol.Compute_LCM(108,901))
```
